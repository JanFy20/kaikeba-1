````python
# 买火车票
buy_ticket_grammar = '''
买火车票 = 寒暄 , 我想买 数量 日期 上下午 时分 从 原地址 到 目的地址 的车票。
寒暄 = 称谓 打招呼 | 打招呼
称谓 = 先生 | 女士
打招呼 = 您好 | 早上好 | 中午好 | 下午好 | 晚上好
数量 = 数字 张
数字 = 1 | 2 | 3 | 4 | 5 | 6
日期 = 数字 月 数字 日
上下午 = 上午 | 下午
时分 = 数字 点 数字 分
原地址 = 城市
目的地址 = 城市
城市 = 北京 | 南京 | 上海 | 广州 | 深圳 | 武汉 | 重庆 | 哈尔滨
'''
````

````python
import random

def create_grammar(grammar_str, split='=', line_split='\n'):
    grammar = {}
    for line in grammar_str.split(line_split):
        if not line.strip(): continue
        exp, stmt = line.split(split)
        grammar[exp.strip()] = [s.split() for s in stmt.split('|')]
    return grammar


def generate(gram, target):
    if target not in gram:
        return target
    
    expaned = [generate(gram, t) for t in random.choice(gram[target])]
    return ''.join([e if e != '/n' else '\n' for e in expaned if e != 'null'])
````

````python
import pandas as pd
import jieba
import re
from collections import Counter

def token(string):
    return re.findall('\w+', string)
````

````python
df = pd.read_csv('movie_comments.csv')
df.head()
````

````python
TOKEN = []
df.fillna('', inplace=True)
articles = df['comment'].tolist()
print('number of articels: %d'%len(articles))
for i in range(len(articles)):
    if i % 10000 == 0: print(i)
    TOKEN += jieba.cut(''.join(token(articles[i])))
````

````python
for i, line in enumerate((open('train.txt', encoding='utf8'))):
    if i % 1000 == 0: print(i)
    TOKEN += jieba.cut(''.join(token(line)))
````

````python
words_count = Counter(TOKEN)
words_count.most_common(20)
````

````python
TOKEN_2_GRAM = [''.join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-2]))]
words_count2 = Counter(TOKEN_2_GRAM)
words_count2.most_common(20)
````

````python
def prob_2(word1, word2):
    if word1 in words_count:
        if word1 + word2 in words_count2: 
            return words_count2[word1+word2] / words_count[word1]
        else:
            return 1 / words_count[word1]
    else:
        return 1/ len(words_count)
    
def get_probability(sentence):
    words = list(jieba.cut(sentence))
    
    sentence_prob = 1
    
    for i, word in enumerate(words[:-1]):
        word = words[i]
        next_ = words[i+1]
        
        probability = prob_2(word, next_)
        
        sentence_prob *= probability
    
    return sentence_prob
````
